```python
# Area plot to show stacked values over time
days = np.arange(1, 8)
water = [2, 3, 4, 3, 5, 6, 7]
milk = [1, 2, 1, 2, 3, 4, 3]
oil = [3,4,5,6,7,8]

plt.figure(figsize=(8, 4))
plt.stackplot(days, water, milk,oil, labels=['Water', 'Milk','Oil'], colors=['blue', 'lightgreen','red'])
plt.title('Area Plot of Liquid Consumption')
plt.xlabel('Day')
plt.ylabel('Litres')
plt.legend()
plt.show()
```
